#Removing -Outliers-IQR

1.   IQR
2.   Z_SCORE



```python
df1=pd.read_csv(r'/content/loan.csv')
df1.head(3)
```

```python
df1.shape
```

```python
plt.figure(figsize=(15,5))
sns.boxplot(x='CoapplicantIncome',data=df1)
plt.show()
```

```python
q1 = df1['CoapplicantIncome'].quantile(0.25)
q3 = df1['CoapplicantIncome'].quantile(0.75)
q3
```

```python
IQR = q3-q1
```

```python
min_range = q1-(1.5*IQR)
max_range = q3+(1.5*IQR)
min_range,max_range
```

```python
new_df1=df1[df1['CoapplicantIncome']<=max_range]
```

```python
new_df1.shape
```

```python
sns.boxplot(x='CoapplicantIncome',data=new_df1)
plt.show()
```
